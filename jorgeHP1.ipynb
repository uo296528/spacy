Carga el `libro1.txt` que contiene el texto del primer libro de Harry Potter, a partir de la línea 38, para realizar los siguientes ejercicios.

````python
# Ruta al archivo (ajusta si está en otra carpeta)
ruta_archivo = "datos2011\\trunk\libros\J.K. Rowling - Harry Potter 1 - La Piedra Filosofal.txt"

# Leer el archivo desde la línea 38 en adelante
with open(ruta_archivo, "r", encoding="utf-8") as f:
    lineas = f.readlines()
    texto = "".join(lineas[37:])  # índice 37 = línea 38 (empieza desde 0)

print(texto[:1000])  # Muestra los primeros 1000 caracteres para verificar
````

````python
import spacy
nlp = spacy.load("es_core_news_md")
doc = nlp(texto)
````

## Ejercicio 1

Extrae todas las entidades nombradas de tipo `PER` (persona) y muestra la lista de personajes encontrados, seguida de los personajes más mencionados junto con su frecuencia.

````python
pers=[]

for ent in doc.ents:
    if ent.label_=="PER":
        pers.append(ent.text)

print(pers)
````

````python
from collections import Counter

pers=[]

for ent in doc.ents:
    if ent.label_=="PER":
        pers.append(ent.text)

print(pers)
pers_dict=Counter(pers).most_common(10)
print(pers_dict)
````

## Ejercicio 2

Detecta oraciones que parecen ser diálogos, es decir, que comienzan por guión

````python
print("Oraciones que parecen diálogos:")
for sent in doc.sents:
    if sent.text.strip().startswith(""):
        print(sent.text.strip())
````

## Ejercicio 3

Detecta menciones de las casas del colegio hogwarts 'Gryffindor' y 'Slytherin' usando `Matcher`.

````python
from spacy.matcher import Matcher


matcher=Matcher(nlp.vocab)


pattern1=[{"TEXT":"Gryffindor"}]
pattern2=[{"TEXT":"Slytherin"}]

matcher.add("casas", [pattern1,pattern2])

for match_id, start, end in matcher(doc):
    # Imprime en pantalla el nombre en string del patrón y
    # el texto del span encontrado
    print(doc[start:end].sent)
````

## Ejercicio 4

Extrae adjetivos cercanos a entidades de tipo `LOC` para entender cómo se describen los lugares.

````python
print("Adjetivos cercanos a entidades de tipo LOC:")

for ent in doc.ents:
    if ent.label_ == "LOC":
        # Revisamos tokens cercanos dentro de una ventana de ±5 palabras
        start = max(ent.start - 5, 0)
        end = min(ent.end + 5, len(doc))
        ventana = doc[start:end]

        # Buscamos adjetivos
        adjetivos = [token.text for token in ventana if token.pos_ == "ADJ"]
        
        if adjetivos:
            print(f"{ent.text} -> [{', '.join(adjetivos)}]")
````

## Ejercicio 5

Extrae verbos relacionados con las emociones que se listan a continuación y cuenta su frecuencia.

```
emociones = ["temer", "amar", "odiar", "llorar", "reír", "sufrir", "sentir"]
```

````python
from collections import Counter
from spacy.matcher import Matcher

emociones = ["temer", "amar", "odiar", "llorar", "reír", "sufrir", "sentir"]
matcher=Matcher(nlp.vocab)

pattern=[{"LEMMA":{"IN":emociones}}]

matcher.add("verb_emocion", [pattern])

lista=[]

for matcher_id, start, end in matcher(doc):
    lista.append(doc[start:end].text.lower())

print(Counter(lista))
````

## Ejercicio 6

Extrae oraciones que contengan al menos dos entidades nombradas (`PER`, `LOC`, `ORG`).

````python
tipos_interes = {"PER", "LOC", "ORG"}

print("Oraciones con al menos 2 entidades nombradas (PER, LOC, ORG):\n")

for sent in doc.sents:
    entidades = [ent for ent in sent.ents if ent.label_ in tipos_interes]
    if len(entidades) >= 2:
        print(f"- {sent.text.strip()}")
        print("----------------------")
````

## Ejercicio 7

Busca oraciones que contengan palabras explicativas, como 'como', 'parecía', 'tal que'.

````python
explica=["como", "parecía", "tal que"]

matcher=Matcher(nlp.vocab)

pattern=[{"LEMMA":{"IN":explica}}]

matcher.add("explicacion", [pattern])


for match_id, start, end in matcher(doc):
    span = doc[start:end]
    oracion = span.sent.text.strip()
    print(f"- {oracion}")
    
````

## Ejercicio 8

Con el modelo que estamos utilizando, los objetos de tipo `Token` a través de su atributo `morph` presentan información morfológica. En concreto, los verbos muestran (con más o menos precisión) entre otras cosas información sobre su tiempo verbal. Teniendo esto en cuenta cuántos verbos hay en cada tiempo verbal usando `token.tag_`.

````python
mini_texto = "Ayer comí pizza. Hoy como con mi hermana, vamos a comer pescado."
mini_doc = nlp(mini_texto)
for token in mini_doc:
    print(f"{token} ---> {token.morph if token.morph else 'No tiene morfología'}")
    # Para sacar un atributo concreto de un objeto de tipo MorphAnalysis haremos ______.get("Tense")
````

````python
# Extraer tiempos verbales desde las características morfológicas
tiempos = [
    token.morph.get("Tense")[0]  # extrae el tiempo verbal, ej. "Pres"
    for token in doc
    if token.pos_ == "VERB" and token.morph.get("Tense")
]

conteo = Counter(tiempos)

print("Conteo de verbos por tiempo verbal (usando token.morph):\n")
for tiempo, freq in conteo.items():
    print(f"- {tiempo}: {freq}")
````

## Ejercicio 9

Utiliza un `PhraseMatcher` para detectar menciones específicas dentro de un fragmento de texto basado en Harry Potter y la piedra filosofal. Las frases que se quieren detectar son expresiones mágicas conocidas como:
- piedra filosofal
- capa invisible

Usa `attr="LOWER"` en el `PhraseMatcher` para que las coincidencias no sean sensibles a mayúsculas

````python
from spacy.matcher import PhraseMatcher

matcher = PhraseMatcher(nlp.vocab)

pattern1 = nlp("piedra filosofal")
pattern2 = nlp("capa invisible")
matcher.add("Ejercicio", [pattern1,pattern2])
int=0
# Itera sobre los resultados
for match_id, start, end in matcher(doc):
    # Obtén el span resultante
    int+=1
    span = doc[start:end]
    print(f"COINCIDENCIA {int}:", span.sent)
    print("--------------------------------")
````

# Ejercicio 10

Utiliza expresiones regulares para localizar los encabezados de los capítulos. Cada capítulo está definido por un número (`1`, `2`, etc.) y un título. Además, te serán últiles las separaciones de las líneas y los espacios en blanco a la hora de definir la expresión regular. Con esta expresión definida, crear un objeto `Span` para cada capítulo utilizando `doc.char_span(...)`.

Definir tres atributos personalizados sobre cada `Span` de capítulo:
    - `.numero`: número del capítulo (como entero).
    - `.titulo`: título del capítulo (como cadena).
    - `.longitud`: longitud del capítulo en número de tokens.

````python
import re
from spacy.tokens import Span
import pandas as pd
Span.set_extension("numero", default=None, force=True)
Span.set_extension("titulo", default=None, force=True)
Span.set_extension("longitud", default=None, force=True)

patron_capitulo = re.finditer(r"\n+(\d+)\s*\n+([^\n]+)", texto, re.MULTILINE)



capitulos = []


coincidencias = list(patron_capitulo)
for i, match in enumerate(coincidencias):
    numero = match.group(1)
    titulo = match.group(2)

    start_char = match.start()
    end_char = coincidencias[i + 1].start() if i + 1 < len(coincidencias) else len(texto)

    span = doc.char_span(start_char, end_char, alignment_mode="expand")

    if span:
        span._.numero = numero
        span._.titulo = titulo
        span._.longitud = len(span)

        capitulos.append(span)


df = pd.DataFrame([{
    "numero": cap._.numero,
    "titulo": cap._.titulo,
    "longitud_tokens": cap._.longitud,
    "texto": cap.text.strip()
} for cap in capitulos])


df
````
